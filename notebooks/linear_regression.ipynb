## Linear Regression Mathematics

**Hypothesis:** h(x) = wx + b

**Cost Function (MSE):**
J(w,b) = (1/2m) * Σ(h(x_i) - y_i)²

**Gradient Descent:**
w := w - α * (∂J/∂w)
b := b - α * (∂J/∂b)

**Gradients:**
∂J/∂w = (1/m) * Σ(h(x_i) - y_i) * x_i
∂J/∂b = (1/m) * Σ(h(x_i) - y_i)

```python
# Calculate cost for specific w, b by hand
X_sample = np.array([1, 2, 3])
y_sample = np.array([3, 5, 7])

w_guess = 2
b_guess = 1

# Predictions
predictions = w_guess * X_sample + b_guess
print(f"Predictions: {predictions}")
print(f"Actual: {y_sample}")

# Cost
errors = predictions - y_sample
cost = (1/(2*len(X_sample))) * np.sum(errors**2)
print(f"Cost: {cost}")

# Gradient
grad_w = (1/len(X_sample)) * np.sum(errors * X_sample)
grad_b = (1/len(X_sample)) * np.sum(errors)
print(f"Gradient w: {grad_w}, Gradient b: {grad_b}")
```
